```python
import yfinance as yf
import pandas as pd

# Example dictionary of stock names and their symbols
symbols = {'SKYGOL': 'SKYGOLD.NS'}

result_df = pd.DataFrame()

for name, symbol in symbols.items():
    stock = yf.Ticker(symbol)
    historical_data = stock.history(period="1y")  # Last 1 year
    historical_data = historical_data.sort_index(ascending=False)  # Sort in descending order

    # Calculate EMAs for 25, 50, 100, 200, and 250 days
    ema_periods = [25, 50, 100, 200, 250]
    for period in ema_periods:
        column_name = f"{period}_day_EMA"
        historical_data[column_name] = historical_data['Close'].ewm(span=period, adjust=False).mean()

    # Extract today's data
    today_data = historical_data.iloc[0]
    
    # Extract data 50 days back
    data_50_days_back = historical_data.iloc[50] if len(historical_data) > 50 else None
    
    # Prepare data for appending
    new_row = {
        "Name": name,
        "Symbol": symbol,
        "Today's Close": today_data['Close'],
        "Today's 25-day EMA": today_data['25_day_EMA'],
        "Today's 50-day EMA": today_data['50_day_EMA'],
        "Today's 100-day EMA": today_data['100_day_EMA'],
        "Today's 200-day EMA": today_data['200_day_EMA'],
        "Today's 250-day EMA": today_data['250_day_EMA'],
    }
    
    if data_50_days_back is not None:
        new_row.update({
            "50 Days Back Close": data_50_days_back['Close'],
            "50 Days Back 25-day EMA": data_50_days_back['25_day_EMA'],
            "50 Days Back 50-day EMA": data_50_days_back['50_day_EMA'],
            "50 Days Back 100-day EMA": data_50_days_back['100_day_EMA'],
            "50 Days Back 200-day EMA": data_50_days_back['200_day_EMA'],
            "50 Days Back 250-day EMA": data_50_days_back['250_day_EMA'],
        })
    
    # Append the data to the result dataframe
    result_df = pd.concat([result_df, pd.DataFrame([new_row])], ignore_index=True)
    
    print(f"{name} ({symbol})")
    print("\nToday's Data:")
    print(today_data)
    
    if data_50_days_back is not None:
        print("\nData 50 Days Back:")
        print(data_50_days_back)

# Display the final dataframe
print(result_df)
```
